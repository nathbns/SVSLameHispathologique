```python
from datasets import Dataset, DatasetDict, Features, Image
import pandas as pd
from pathlib import Path
BASE_DIR = Path("../dataset_patches_256")
META = pd.read_csv(BASE_DIR / "patches_metadata.csv")

def make_split(split_name):
    df = META[META["split"] == split_name].copy()
    df["image"] = df["filename"].apply(lambda f: str(BASE_DIR / split_name / f))
    return Dataset.from_pandas(
        df[["image", "patch_id", "slide_name", "x", "y", "split", "tissue_ratio"]]
    ).cast_column("image", Image())
ds = DatasetDict({
    "train": make_split("train"),
    "validation": make_split("val"),
})
ds.push_to_hub("nathbns/SVS-TCGA-BR")
```
